```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def get_input_color():
    myColor = [200,120,50]
    return myColor

def find_adjacent_color(rgb):
    hsv = rgb_to_hsv(rgb)
    a = hsv_to_rgb([hsv[0][0][0]+30,hsv[0][0][1],hsv[0][0][2]])
    b = hsv_to_rgb([hsv[0][0][0]-30,hsv[0][0][1],hsv[0][0][2]])
    return a,b
```

```python
# h:0-360 , s:0-255, v:0-255
# r:0-255, g:0-255, b:0-255
def rgb_to_hsv(rgb=[]):
    img=np.array([[rgb]],np.uint8)
    img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
    img_hsv[0][0]*=2
    return img_hsv

def hsv_to_rgb(hsv=[]):
    hsv[0]/=2
    img_hsv=np.array([[hsv]],np.uint8)
    img_rgb = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB)
    return img_rgb
```

```python
# 1.获取颜色
colorInput = get_input_color()

# 2. 寻找邻近色
color1,color2 = find_adjacent_color(colorInput)

# 3. 输出配色方案
print("input color:",colorInput)
img1=np.array([[colorInput]],np.uint8)
plt.imshow(img1)
plt.show()
print("color1:",color1[0][0])
#img2=np.array([[color1]],np.uint8)
plt.imshow(color1)
plt.show()
print("color2:",color2[0][0])
#img3=np.array([[color2]],np.uint8)
plt.imshow(color2)
plt.show()
```
